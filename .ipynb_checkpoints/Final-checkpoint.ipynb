```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2

# ================== CONFIG ==================
file_path = 'NP-05_SCN0001.txt'
z_target  = 13.6124
dz        = 1.0
max_points_for_speed = 1_000_000
inlier_quantile = 0.80    # giữ 80% điểm tốt nhất khi refit

# --------- Circle fit (Pratt) ----------
def fit_circle_pratt(x, y):
    x = np.asarray(x, dtype=float)
    y = np.asarray(y, dtype=float)
    xm, ym = x.mean(), y.mean()
    u, v = x - xm, y - ym
    Suu, Suv, Svv = np.sum(u*u), np.sum(u*v), np.sum(v*v)
    Suuu, Svvv = np.sum(u*u*u), np.sum(v*v*v)
    Suvv, Svuu = np.sum(u*v*v), np.sum(v*u*u)
    A = np.array([[Suu, Suv],[Suv, Svv]], float)
    b = 0.5*np.array([Suuu + Suvv, Svvv + Svuu], float)
    try:
        uc, vc = np.linalg.solve(A, b)
    except np.linalg.LinAlgError:
        uc, vc = np.linalg.pinv(A) @ b
    xc, yc = xm + uc, ym + vc
    R = np.sqrt(uc*uc + vc*vc + (Suu+Svv)/len(x))
    return xc, yc, R

# ================== 1) Đọc file ROBUST ==================
# - engine='python' cho phép regex separator
# - sep=r'[,\s;]+' chấp nhận comma / space / tab / ;
# - on_bad_lines='skip' bỏ các dòng có chữ như 'X'
df = pd.read_csv(
    file_path,
    header=None,
    names=['x', 'y', 'z'],
    sep=r'[,\s;]+',
    engine='python',
    usecols=[0,1,2],
    comment='#',
    skip_blank_lines=True,
    on_bad_lines='skip',
    dtype=str  # đọc thô dạng chuỗi, rồi ép số an toàn phía dưới
)

# Ép kiểu số, bỏ dòng lỗi
for c in ['x','y','z']:
    df[c] = pd.to_numeric(df[c].str.strip(), errors='coerce')
df = df.dropna()

# ================== 2) Lát cắt ==================
mask = (df['z'] >= z_target - dz) & (df['z'] <= z_target + dz)
slc = df.loc[mask, ['x','y']].to_numpy()
n0 = len(slc)
if n0 < 1000:
    raise RuntimeError(f"Lát cắt quá thưa ({n0} điểm). Tăng dz hoặc chọn z khác.")
if n0 > max_points_for_speed:
    idx = np.random.choice(n0, max_points_for_speed, replace=False)
    slc = slc[idx]
print("Số điểm lát cắt dùng:", len(slc))

# ================== 3) Convex Hull trực tiếp ==================
pts32 = slc.astype(np.float32).reshape(-1,1,2)
hull = cv2.convexHull(pts32, returnPoints=True)
hull_xy = hull.reshape(-1,2).astype(np.float64)
if len(hull_xy) < 10:
    raise RuntimeError("Convex hull quá ít đỉnh; biên ngoài có thể bị khuyết.")

# ================== 4) Pratt fit + refit robust ==================
xc, yc, R = fit_circle_pratt(hull_xy[:,0], hull_xy[:,1])

rho = np.sqrt((hull_xy[:,0]-xc)**2 + (hull_xy[:,1]-yc)**2)
res = np.abs(rho - R)
thr = np.quantile(res, inlier_quantile)  # giữ 80% điểm tốt nhất
good = res <= thr
xc2, yc2, R2 = fit_circle_pratt(hull_xy[good,0], hull_xy[good,1])

print("\n--- Kết quả (ConvexHull → Pratt) ---")
print(f"Center (x,y): ({xc2:.6f}, {yc2:.6f})")
print(f"Radius      : {R2:.6f}")

# ================== 5) Visualize ==================
theta = np.linspace(0, 2*np.pi, 720)
circ_x = xc2 + R2*np.cos(theta)
circ_y = yc2 + R2*np.sin(theta)

fig, ax = plt.subplots(1,1, figsize=(6,6))
ax.scatter(slc[:,0], slc[:,1], s=0.2, alpha=0.2, label="slice points")
ax.plot(hull_xy[:,0], hull_xy[:,1], 'k-', linewidth=1, alpha=0.9, label="convex hull")
ax.plot(circ_x, circ_y, linewidth=2, label="fitted circle")
ax.scatter([xc2], [yc2], s=25, label="center")
ax.set_aspect('equal', 'box'); ax.grid(True); ax.legend()
ax.set_title(f"z={z_target} ± {dz} | circle via convex hull")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ================== CONFIG ==================
file_path = 'NP-05_SCN0001.txt'
z_target  = 21.1314    # z trung tâm lát cắt
dz        = 5.0        # độ dày lát cắt (cùng đơn vị với dữ liệu)
R_known   = 3.0     # BÁN KÍNH CỐ ĐỊNH (ví dụ 3 m nếu dữ liệu là mm -> 3000.0)

# LM / Robust
use_huber = True
huber_c   = 2.5
max_iter  = 1000
lam0      = 1e-2
tol_step  = 1e-9
tol_cost  = 1e-12

# ================== HÀM PHỤ ==================
def taubin_circle_fit(x, y):
    """
    Circle fit by Taubin (Chernov-Lesort). Trả về (xc, yc, R).
    Dùng làm khởi tạo tâm tốt, sau đó R sẽ bị cố định ở bước LM.
    """
    x = np.asarray(x, float); y = np.asarray(y, float)
    xm, ym = x.mean(), y.mean()
    u, v   = x - xm, y - ym
    Suu = np.sum(u*u); Suv = np.sum(u*v); Svv = np.sum(v*v)
    Suuu = np.sum(u*u*u); Svvv = np.sum(v*v*v)
    Suvv = np.sum(u*v*v); Svuu = np.sum(v*u*u)

    A = np.array([[Suu, Suv],
                  [Suv, Svv]], float)
    B = 0.5 * np.array([Suuu + Suvv,
                        Svvv + Svuu], float)
    try:
        uc, vc = np.linalg.solve(A, B)
    except np.linalg.LinAlgError:
        uc, vc = np.linalg.pinv(A) @ B

    xc, yc = xm + uc, ym + vc
    R = np.sqrt(uc*uc + vc*vc + (Suu + Svv)/len(x))
    return xc, yc, R

def huber_weights(r, c):
    """
    Trọng số Huber cho residual r (chuẩn hoá theo sigma).
    """
    a = np.abs(r)
    w = np.ones_like(a)
    m = a > c
    w[m] = c / a[m]
    return w

def lm_center_with_fixed_R(x, y, xc0, yc0, R_fixed,
                           use_huber=True, huber_c=2.5,
                           max_iter=100, lam0=1e-2, tol_step=1e-9, tol_cost=1e-12):
    """
    LM tối thiểu hoá sum w_i * ( sqrt((x-xc)^2+(y-yc)^2 ) - R_fixed )^2
    Biến số: theta = [xc, yc], R cố định.
    """
    x = np.asarray(x, float); y = np.asarray(y, float)
    n = x.size
    xc, yc = float(xc0), float(yc0)
    R = float(R_fixed)
    lam = lam0

    # cost ban đầu
    dx = x - xc; dy = y - yc
    ri = np.sqrt(dx*dx + dy*dy) - R
    if use_huber:
        sigma = 1.4826 * np.median(np.abs(ri - np.median(ri)))
        sigma = sigma if sigma > 1e-12 else np.sqrt(np.mean(ri**2) + 1e-12)
        w = huber_weights(ri / (sigma + 1e-12), huber_c)
    else:
        w = np.ones_like(ri)
    cost = np.sum(w * ri * ri)

    for it in range(max_iter):
        dx = x - xc; dy = y - yc
        d  = np.sqrt(dx*dx + dy*dy) + 1e-15
        ri = d - R

        if use_huber:
            sigma = 1.4826 * np.median(np.abs(ri - np.median(ri)))
            sigma = sigma if sigma > 1e-12 else np.sqrt(np.mean(ri**2) + 1e-12)
            w = huber_weights(ri / (sigma + 1e-12), huber_c)
        else:
            w = np.ones_like(ri)

        # Jacobian theo [xc, yc]
        J = np.empty((n,2), dtype=float)
        J[:,0] = -dx / d
        J[:,1] = -dy / d

        JW = J * w[:,None]
        H = JW.T @ J      # 2x2
        g = JW.T @ ri     # 2

        H_lm = H + lam * np.eye(2)
        try:
            delta = -np.linalg.solve(H_lm, g)
        except np.linalg.LinAlgError:
            delta = -np.linalg.pinv(H_lm) @ g

        xc_new = xc + delta[0]
        yc_new = yc + delta[1]

        # cost mới
        dxn = x - xc_new; dyn = y - yc_new
        dn  = np.sqrt(dxn*dxn + dyn*dyn) + 1e-15
        rin = dn - R
        if use_huber:
            sigma_n = 1.4826 * np.median(np.abs(rin - np.median(rin)))
            sigma_n = sigma_n if sigma_n > 1e-12 else np.sqrt(np.mean(rin**2) + 1e-12)
            wn = huber_weights(rin / (sigma_n + 1e-12), huber_c)
        else:
            wn = np.ones_like(rin)
        cost_new = np.sum(wn * rin * rin)

        if cost_new < cost:
            xc, yc = xc_new, yc_new
            if np.linalg.norm(delta) < tol_step or abs(cost - cost_new) < tol_cost:
                cost = cost_new
                break
            cost = cost_new
            lam = max(lam * 0.3, 1e-12)
        else:
            lam = min(lam * 5.0, 1e12)

    return xc, yc, R, cost, it+1

# ================== ĐỌC FILE (robust) ==================
df = pd.read_csv(
    file_path,
    header=None,
    names=['x','y','z'],
    sep=r'[,\s;]+',
    engine='python',
    usecols=[0,1,2],
    comment='#',
    skip_blank_lines=True,
    on_bad_lines='skip',
    dtype=str
)
for c in ['x','y','z']:
    df[c] = pd.to_numeric(df[c].str.strip(), errors='coerce')
df = df.dropna()

# Thông tin z để chọn lát cắt hợp lý
print(f"z range: [{df['z'].min():.6f}, {df['z'].max():.6f}] | total pts: {len(df)}")

# ================== LỌC LÁT CẮT ==================
mask = (df['z'] >= z_target - dz) & (df['z'] <= z_target + dz)
xy = df.loc[mask, ['x','y']].to_numpy(dtype=float)
n_slice = len(xy)
print(f"Slice @ z={z_target} ± {dz}: {n_slice} points")
if n_slice < 50:
    raise RuntimeError("Lát cắt quá thưa; tăng dz hoặc chọn z khác.")

x, y = xy[:,0], xy[:,1]

# ================== TAUBIN (init) → LM (R cố định) ==================
xc0, yc0, R0 = taubin_circle_fit(x, y)  # R0 chỉ để tham khảo
xc, yc, R, cost, iters = lm_center_with_fixed_R(
    x, y, xc0, yc0, R_known,
    use_huber=use_huber, huber_c=huber_c,
    max_iter=max_iter, lam0=lam0, tol_step=tol_step, tol_cost=tol_cost
)

print("\n=== Circle fit: R fixed ===")
print(f"R (fixed)    : {R:.6f}")
print(f"Init center  : ({xc0:.6f}, {yc0:.6f})  [Taubin],  R0(Taubin)={R0:.6f}")
print(f"LM center    : ({xc:.6f}, {yc:.6f})  | iters: {iters} | cost: {cost:.6e}")

# ================== PLOT ==================
theta = np.linspace(0, 2*np.pi, 720)
xcirc = xc + R*np.cos(theta)
ycirc = yc + R*np.sin(theta)

rho = np.sqrt((x-xc)**2 + (y-yc)**2)
res = rho - R
sd  = np.std(res)
inlier = np.abs(res) <= 2.5*sd if sd>0 else np.ones_like(res, bool)

fig, ax = plt.subplots(1,1, figsize=(6,6))
ax.scatter(x[~inlier], y[~inlier], s=1, alpha=0.25, label='outliers')
ax.scatter(x[inlier],  y[inlier],  s=1, alpha=0.85, label='inliers')
ax.plot(xcirc, ycirc, linewidth=2, label=f'circle (R fixed = {R_known})')
ax.scatter([xc],[yc], s=25, label='center')
ax.set_aspect('equal', 'box'); ax.grid(True); ax.legend()
ax.set_title(f"z={z_target} ± {dz}\ncenter=({xc:.4f},{yc:.4f}), R={R:.4f}")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ================== CONFIG ==================
file_path = 'NP-05_SCN0001.txt'
z_target  = 21.1314    # z trung tâm lát cắt
dz        = 5.0        # độ dày lát cắt (cùng đơn vị với dữ liệu)
R_known   = 3.0        # BÁN KÍNH CỐ ĐỊNH (ví dụ: 3.0)

# RANSAC
ransac_max_trials = 5000
inlier_thresh     = 0.50   # ngưỡng residual (cùng đơn vị với dữ liệu)
min_inliers_frac  = 0.25   # tối thiểu % inlier để chấp nhận model
random_seed       = 42

# (Tuỳ chọn) refine tâm bằng LM sau khi RANSAC khoanh inliers
do_lm_refine = True
use_huber = True
huber_c   = 2.5
max_iter  = 500
lam0      = 1e-2
tol_step  = 1e-9
tol_cost  = 1e-12

# ================== HÀM PHỤ ==================
def huber_weights(r, c):
    a = np.abs(r)
    w = np.ones_like(a)
    m = a > c
    w[m] = c / a[m]
    return w

def lm_center_with_fixed_R(x, y, xc0, yc0, R_fixed,
                           use_huber=True, huber_c=2.5,
                           max_iter=100, lam0=1e-2, tol_step=1e-9, tol_cost=1e-12):
    """
    LM tối thiểu hoá sum w_i * ( sqrt((x-xc)^2+(y-yc)^2 ) - R_fixed )^2
    Biến số: [xc, yc], R cố định.
    """
    x = np.asarray(x, float); y = np.asarray(y, float)
    n = x.size
    xc, yc = float(xc0), float(yc0)
    R = float(R_fixed)
    lam = lam0

    # cost ban đầu
    dx = x - xc; dy = y - yc
    ri = np.sqrt(dx*dx + dy*dy) - R
    if use_huber:
        sigma = 1.4826 * np.median(np.abs(ri - np.median(ri)))
        sigma = sigma if sigma > 1e-12 else np.sqrt(np.mean(ri**2) + 1e-12)
        w = huber_weights(ri / (sigma + 1e-12), huber_c)
    else:
        w = np.ones_like(ri)
    cost = np.sum(w * ri * ri)

    for it in range(max_iter):
        dx = x - xc; dy = y - yc
        d  = np.sqrt(dx*dx + dy*dy) + 1e-15
        ri = d - R

        if use_huber:
            sigma = 1.4826 * np.median(np.abs(ri - np.median(ri)))
            sigma = sigma if sigma > 1e-12 else np.sqrt(np.mean(ri**2) + 1e-12)
            w = huber_weights(ri / (sigma + 1e-12), huber_c)
        else:
            w = np.ones_like(ri)

        # Jacobian theo [xc, yc]
        J = np.empty((n,2), dtype=float)
        J[:,0] = -dx / d
        J[:,1] = -dy / d

        JW = J * w[:,None]
        H = JW.T @ J      # 2x2
        g = JW.T @ ri     # 2

        H_lm = H + lam * np.eye(2)
        try:
            delta = -np.linalg.solve(H_lm, g)
        except np.linalg.LinAlgError:
            delta = -np.linalg.pinv(H_lm) @ g

        xc_new = xc + delta[0]
        yc_new = yc + delta[1]

        # cost mới
        dxn = x - xc_new; dyn = y - yc_new
        dn  = np.sqrt(dxn*dxn + dyn*dyn) + 1e-15
        rin = dn - R
        if use_huber:
            sigma_n = 1.4826 * np.median(np.abs(rin - np.median(rin)))
            sigma_n = sigma_n if sigma_n > 1e-12 else np.sqrt(np.mean(rin**2) + 1e-12)
            wn = huber_weights(rin / (sigma_n + 1e-12), huber_c)
        else:
            wn = np.ones_like(rin)
        cost_new = np.sum(wn * rin * rin)

        if cost_new < cost:
            xc, yc = xc_new, yc_new
            if np.linalg.norm(delta) < tol_step or abs(cost - cost_new) < tol_cost:
                cost = cost_new
                break
            cost = cost_new
            lam = max(lam * 0.3, 1e-12)
        else:
            lam = min(lam * 5.0, 1e12)

    return xc, yc, R, cost, it+1

def centers_from_two_points_fixed_R(p1, p2, R):
    """
    Trả về tối đa 2 tâm (x1,y1),(x2,y2) của đường tròn bán kính R đi qua p1,p2.
    Nếu khoảng cách > 2R: không có nghiệm (return []).
    Nếu khoảng cách ~ 2R: 1 nghiệm (tiếp xúc).
    """
    (x1,y1), (x2,y2) = np.asarray(p1,float), np.asarray(p2,float)
    dx, dy = x2 - x1, y2 - y1
    d = np.hypot(dx, dy)
    if d < 1e-15 or d > 2*R + 1e-15:
        return []  # trùng điểm hoặc quá xa -> vô nghiệm cho R cố định
    # trung điểm
    mx, my = (x1 + x2)/2.0, (y1 + y2)/2.0
    # khoảng cách từ trung điểm đến tâm theo phương vuông góc
    h_sq = R*R - (d*d)/4.0
    if h_sq < 0:  # phòng số học
        return []
    h = np.sqrt(max(h_sq, 0.0))
    # vectơ đơn vị vuông góc với (dx,dy)
    ux, uy = -dy/d, dx/d
    c1 = (mx + h*ux, my + h*uy)
    if h < 1e-15:
        return [c1]  # tiếp xúc -> 1 tâm
    c2 = (mx - h*ux, my - h*uy)
    return [c1, c2]

def consensus_score(x, y, xc, yc, R, inlier_thresh):
    rho = np.sqrt((x - xc)**2 + (y - yc)**2)
    res = np.abs(rho - R)
    inliers = res <= inlier_thresh
    return inliers, res

def ransac_circle_fixed_R(x, y, R, inlier_thresh=0.5, max_trials=2000,
                          min_inliers_frac=0.25, random_seed=42):
    """
    RANSAC với mô hình: tâm (xc,yc), R cố định.
    Mẫu: 2 điểm → 0/1/2 tâm ứng viên. Chọn ứng viên cho inlier nhiều nhất.
    """
    x = np.asarray(x,float); y = np.asarray(y,float)
    n = x.size
    if n < 10:
        raise RuntimeError("Quá ít điểm cho RANSAC")

    rng = np.random.default_rng(random_seed)
    best_inliers = None
    best_center  = None
    best_support = -1

    idx = np.arange(n)

    for _ in range(max_trials):
        i1, i2 = rng.choice(idx, size=2, replace=False)
        p1 = (x[i1], y[i1])
        p2 = (x[i2], y[i2])
        centers = centers_from_two_points_fixed_R(p1, p2, R)
        if not centers:
            continue
        for (xc, yc) in centers:
            inliers, _ = consensus_score(x, y, xc, yc, R, inlier_thresh)
            support = np.count_nonzero(inliers)
            if support > best_support:
                best_support = support
                best_inliers = inliers
                best_center  = (xc, yc)

    if best_center is None:
        raise RuntimeError("RANSAC không tìm được mô hình phù hợp (R quá nhỏ/lớn hoặc dữ liệu nhiễu)")

    if best_support < max(int(min_inliers_frac * n), 5):
        raise RuntimeError(f"RANSAC tìm thấy {best_support} inliers (< ngưỡng yêu cầu). Hãy nới inlier_thresh hoặc tăng dz / chọn z khác.")

    return best_center, best_inliers

# ================== ĐỌC FILE (robust) ==================
df = pd.read_csv(
    file_path,
    header=None,
    names=['x','y','z'],
    sep=r'[,\s;]+',
    engine='python',
    usecols=[0,1,2],
    comment='#',
    skip_blank_lines=True,
    on_bad_lines='skip',
    dtype=str
)
for c in ['x','y','z']:
    df[c] = pd.to_numeric(df[c].str.strip(), errors='coerce')
df = df.dropna()

# Thông tin z để chọn lát cắt
print(f"z range: [{df['z'].min():.6f}, {df['z'].max():.6f}] | total pts: {len(df)}")

# ================== LỌC LÁT CẮT ==================
mask = (df['z'] >= z_target - dz) & (df['z'] <= z_target + dz)
xy = df.loc[mask, ['x','y']].to_numpy(dtype=float)
n_slice = len(xy)
print(f"Slice @ z={z_target} ± {dz}: {n_slice} points")
if n_slice < 50:
    raise RuntimeError("Lát cắt quá thưa; tăng dz hoặc chọn z khác.")

x, y = xy[:,0], xy[:,1]

# ================== RANSAC (R cố định) ==================
(best_xc, best_yc), inliers_mask = ransac_circle_fixed_R(
    x, y, R_known,
    inlier_thresh=inlier_thresh,
    max_trials=ransac_max_trials,
    min_inliers_frac=min_inliers_frac,
    random_seed=random_seed
)
print("\n=== RANSAC (R fixed) ===")
print(f"Center (RANSAC): ({best_xc:.6f}, {best_yc:.6f})")
print(f"Inliers        : {inliers_mask.sum()} / {len(x)} (threshold={inlier_thresh})")

# (Tuỳ chọn) refine tâm bằng LM chỉ trên inliers
if do_lm_refine:
    xin, yin = x[inliers_mask], y[inliers_mask]
    xc, yc, R, cost, iters = lm_center_with_fixed_R(
        xin, yin, best_xc, best_yc, R_known,
        use_huber=use_huber, huber_c=huber_c,
        max_iter=max_iter, lam0=lam0, tol_step=tol_step, tol_cost=tol_cost
    )
else:
    xc, yc, R = best_xc, best_yc, R_known
    # ước lượng cost đơn giản
    rho_tmp = np.sqrt((x[inliers_mask]-xc)**2 + (y[inliers_mask]-yc)**2)
    res_tmp = rho_tmp - R
    cost = np.sum(res_tmp**2)
    iters = 0

print("\n=== Final (R fixed) ===")
print(f"R (fixed)   : {R:.6f}")
print(f"Center final: ({xc:.6f}, {yc:.6f}) | LM iters: {iters} | cost: {cost:.6e}")

# ================== PLOT ==================
theta = np.linspace(0, 2*np.pi, 720)
xcirc = xc + R*np.cos(theta)
ycirc = yc + R*np.sin(theta)

rho = np.sqrt((x-xc)**2 + (y-yc)**2)
res = rho - R
sd  = np.std(res)
auto_inlier = np.abs(res) <= 2.5*sd if sd > 0 else np.ones_like(res, bool)

fig, ax = plt.subplots(1,1, figsize=(6,6))
# hiển thị outlier theo RANSAC
ax.scatter(x[~inliers_mask], y[~inliers_mask], s=1, alpha=0.25, label='RANSAC outliers')
ax.scatter(x[inliers_mask],  y[inliers_mask],  s=1, alpha=0.85, label='RANSAC inliers')
ax.plot(xcirc, ycirc, linewidth=2, label=f'circle (R fixed = {R_known})')
ax.scatter([xc],[yc], s=25, label='center')
ax.set_aspect('equal', 'box'); ax.grid(True); ax.legend()
ax.set_title(f"z={z_target} ± {dz}\ncenter=({xc:.4f},{yc:.4f}), R={R:.4f}\n"
             f"inliers={inliers_mask.sum()}/{len(x)} (thr={inlier_thresh})")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ================== CONFIG ==================
file_path = 'NP-05_SCN0001.txt'

# Lựa chọn bán kính:
# - Nếu biết bán kính thiết kế: đặt R_known = 3.0 (ví dụ)
# - Nếu KHÔNG biết: đặt R_known = None -> fit cả (xc,yc,R)
R_known   = None           # None => không biết R, sẽ fit R luôn; số dương => cố định R

dz        = 5.0            # bề dày lát cắt (m)
z_step    = 4.0            # khoảng cách giữa các cao độ (m)

# RANSAC
ransac_max_trials = 5000
inlier_thresh     = 0.50   # ngưỡng residual |rho-R| (m)
min_inliers_frac  = 0.25
random_seed       = 42

# (Tuỳ chọn) refine bằng LM sau khi RANSAC khoanh inliers
do_lm_refine = True
use_huber = True
huber_c   = 2.5
max_iter  = 500
lam0      = 1e-2
tol_step  = 1e-9
tol_cost  = 1e-12

# ================== HÀM PHỤ ==================
def huber_weights(r, c):
    a = np.abs(r)
    w = np.ones_like(a)
    m = a > c
    w[m] = c / (a[m] + 1e-15)
    return w

def lm_center_with_fixed_R(x, y, xc0, yc0, R_fixed,
                           use_huber=True, huber_c=2.5,
                           max_iter=100, lam0=1e-2, tol_step=1e-9, tol_cost=1e-12):
    x = np.asarray(x, float); y = np.asarray(y, float)
    n = x.size
    xc, yc = float(xc0), float(yc0)
    R = float(R_fixed)
    lam = lam0

    dx = x - xc; dy = y - yc
    ri = np.sqrt(dx*dx + dy*dy) - R
    if use_huber:
        sigma = 1.4826 * np.median(np.abs(ri - np.median(ri)))
        sigma = sigma if sigma > 1e-12 else np.sqrt(np.mean(ri**2) + 1e-12)
        w = huber_weights(ri / (sigma + 1e-12), huber_c)
    else:
        w = np.ones_like(ri)
    cost = np.sum(w * ri * ri)

    for it in range(max_iter):
        dx = x - xc; dy = y - yc
        d  = np.sqrt(dx*dx + dy*dy) + 1e-15
        ri = d - R

        if use_huber:
            sigma = 1.4826 * np.median(np.abs(ri - np.median(ri)))
            sigma = sigma if sigma > 1e-12 else np.sqrt(np.mean(ri**2) + 1e-12)
            w = huber_weights(ri / (sigma + 1e-12), huber_c)
        else:
            w = np.ones_like(ri)

        # Jacobian theo [xc, yc]
        J = np.empty((n,2), dtype=float)
        J[:,0] = -dx / d
        J[:,1] = -dy / d

        JW = J * w[:,None]
        H = JW.T @ J
        g = JW.T @ ri

        H_lm = H + lam * np.eye(2)
        try:
            delta = -np.linalg.solve(H_lm, g)
        except np.linalg.LinAlgError:
            delta = -np.linalg.pinv(H_lm) @ g

        xc_new = xc + delta[0]
        yc_new = yc + delta[1]

        dxn = x - xc_new; dyn = y - yc_new
        dn  = np.sqrt(dxn*dxn + dyn*dyn) + 1e-15
        rin = dn - R
        if use_huber:
            sigma_n = 1.4826 * np.median(np.abs(rin - np.median(rin)))
            sigma_n = sigma_n if sigma_n > 1e-12 else np.sqrt(np.mean(rin**2) + 1e-12)
            wn = huber_weights(rin / (sigma_n + 1e-12), huber_c)
        else:
            wn = np.ones_like(rin)
        cost_new = np.sum(wn * rin * rin)

        if cost_new < cost:
            xc, yc = xc_new, yc_new
            if np.linalg.norm(delta) < tol_step or abs(cost - cost_new) < tol_cost:
                cost = cost_new
                break
            cost = cost_new
            lam = max(lam * 0.3, 1e-12)
        else:
            lam = min(lam * 5.0, 1e12)

    return xc, yc, R, cost, it+1

def lm_center_radius(x, y, xc0, yc0, R0,
                     use_huber=True, huber_c=2.5,
                     max_iter=100, lam0=1e-2, tol_step=1e-9, tol_cost=1e-12):
    """
    LM tối thiểu hoá sum w_i * (sqrt((x-xc)^2+(y-yc)^2) - R)^2
    Biến số: [xc, yc, R]
    """
    x = np.asarray(x, float); y = np.asarray(y, float)
    n = x.size
    xc, yc, R = float(xc0), float(yc0), float(R0)
    lam = lam0

    dx = x - xc; dy = y - yc
    d  = np.sqrt(dx*dx + dy*dy) + 1e-15
    ri = d - R
    if use_huber:
        sigma = 1.4826 * np.median(np.abs(ri - np.median(ri)))
        sigma = sigma if sigma > 1e-12 else np.sqrt(np.mean(ri**2) + 1e-12)
        w = huber_weights(ri / (sigma + 1e-12), huber_c)
    else:
        w = np.ones_like(ri)
    cost = np.sum(w * ri * ri)

    for it in range(max_iter):
        dx = x - xc; dy = y - yc
        d  = np.sqrt(dx*dx + dy*dy) + 1e-15
        ri = d - R

        if use_huber:
            sigma = 1.4826 * np.median(np.abs(ri - np.median(ri)))
            sigma = sigma if sigma > 1e-12 else np.sqrt(np.mean(ri**2) + 1e-12)
            w = huber_weights(ri / (sigma + 1e-12), huber_c)
        else:
            w = np.ones_like(ri)

        # Jacobian theo [xc, yc, R]
        J = np.empty((n,3), dtype=float)
        J[:,0] = -dx / d
        J[:,1] = -dy / d
        J[:,2] = -1.0

        JW = J * w[:,None]
        H = JW.T @ J
        g = JW.T @ ri

        H_lm = H + lam * np.eye(3)
        try:
            delta = -np.linalg.solve(H_lm, g)
        except np.linalg.LinAlgError:
            delta = -np.linalg.pinv(H_lm) @ g

        xc_new = xc + delta[0]
        yc_new = yc + delta[1]
        R_new  = R  + delta[2]

        dxn = x - xc_new; dyn = y - yc_new
        dn  = np.sqrt(dxn*dxn + dyn*dyn) + 1e-15
        rin = dn - R_new
        if use_huber:
            sigma_n = 1.4826 * np.median(np.abs(rin - np.median(rin)))
            sigma_n = sigma_n if sigma_n > 1e-12 else np.sqrt(np.mean(rin**2) + 1e-12)
            wn = huber_weights(rin / (sigma_n + 1e-12), huber_c)
        else:
            wn = np.ones_like(rin)
        cost_new = np.sum(wn * rin * rin)

        if cost_new < cost and R_new > 0:
            xc, yc, R = xc_new, yc_new, R_new
            if np.linalg.norm(delta) < tol_step or abs(cost - cost_new) < tol_cost:
                cost = cost_new
                break
            cost = cost_new
            lam = max(lam * 0.3, 1e-12)
        else:
            lam = min(lam * 5.0, 1e12)

    return xc, yc, R, cost, it+1

def circle_from_3points(p1, p2, p3, eps=1e-12):
    """
    Trả về (xc,yc,R) của đường tròn qua 3 điểm (không thẳng hàng).
    """
    (x1,y1), (x2,y2), (x3,y3) = map(lambda p: (float(p[0]), float(p[1])), (p1,p2,p3))
    A = np.array([[x2-x1, y2-y1],
                  [x3-x1, y3-y1]], dtype=float)
    b = 0.5*np.array([x2*x2 - x1*x1 + y2*y2 - y1*y1,
                      x3*x3 - x1*x1 + y3*y3 - y1*y1], dtype=float)
    if abs(np.linalg.det(A)) < eps:
        return None  # gần thẳng hàng
    try:
        cxcy = np.linalg.solve(A, b)
    except np.linalg.LinAlgError:
        cxcy = np.linalg.pinv(A) @ b
    xc, yc = cxcy[0], cxcy[1]
    R = np.hypot(x1 - xc, y1 - yc)
    if R <= eps:
        return None
    return xc, yc, R

def centers_from_two_points_fixed_R(p1, p2, R):
    (x1,y1), (x2,y2) = np.asarray(p1,float), np.asarray(p2,float)
    dx, dy = x2 - x1, y2 - y1
    d = np.hypot(dx, dy)
    if d < 1e-15 or d > 2*R + 1e-15:  # trùng điểm hoặc quá xa
        return []
    mx, my = (x1 + x2)/2.0, (y1 + y2)/2.0
    h_sq = R*R - (d*d)/4.0
    if h_sq < 0:
        return []
    h = np.sqrt(max(h_sq, 0.0))
    ux, uy = -dy/d, dx/d
    c1 = (mx + h*ux, my + h*uy)
    if h < 1e-15:
        return [c1]
    c2 = (mx - h*ux, my - h*uy)
    return [c1, c2]

def consensus_score(x, y, xc, yc, R, inlier_thresh):
    rho = np.sqrt((x - xc)**2 + (y - yc)**2)
    res = np.abs(rho - R)
    inliers = res <= inlier_thresh
    return inliers, res

def ransac_circle_fixed_R(x, y, R, inlier_thresh=0.5, max_trials=2000,
                          min_inliers_frac=0.25, random_seed=42):
    x = np.asarray(x,float); y = np.asarray(y,float)
    n = x.size
    if n < 10:
        raise RuntimeError("Quá ít điểm cho RANSAC (R cố định)")
    rng = np.random.default_rng(random_seed)
    best_inliers = None
    best_center  = None
    best_support = -1
    idx = np.arange(n)

    for _ in range(max_trials):
        i1, i2 = rng.choice(idx, size=2, replace=False)
        p1 = (x[i1], y[i1])
        p2 = (x[i2], y[i2])
        centers = centers_from_two_points_fixed_R(p1, p2, R)
        if not centers:
            continue
        for (xc, yc) in centers:
            inliers, _ = consensus_score(x, y, xc, yc, R, inlier_thresh)
            support = np.count_nonzero(inliers)
            if support > best_support:
                best_support = support
                best_inliers = inliers
                best_center  = (xc, yc)

    if best_center is None:
        raise RuntimeError("RANSAC (R cố định) không tìm được mô hình.")
    if best_support < max(int(min_inliers_frac * n), 5):
        raise RuntimeError(f"RANSAC tìm thấy {best_support} inliers (< ngưỡng).")

    return best_center, best_inliers

def ransac_circle_variable_R(x, y, inlier_thresh=0.5, max_trials=2000,
                             min_inliers_frac=0.25, random_seed=42):
    """
    RANSAC với mô hình (xc,yc,R) — mẫu tối thiểu 3 điểm không thẳng hàng.
    """
    x = np.asarray(x,float); y = np.asarray(y,float)
    n = x.size
    if n < 15:
        raise RuntimeError("Quá ít điểm cho RANSAC (R tự do)")
    rng = np.random.default_rng(random_seed)
    best_inliers = None
    best_params  = None
    best_support = -1
    idx = np.arange(n)

    for _ in range(max_trials):
        i1, i2, i3 = rng.choice(idx, size=3, replace=False)
        p1 = (x[i1], y[i1]); p2 = (x[i2], y[i2]); p3 = (x[i3], y[i3])
        circ = circle_from_3points(p1,p2,p3)
        if circ is None:
            continue
        xc, yc, R = circ
        inliers, _ = consensus_score(x, y, xc, yc, R, inlier_thresh)
        support = np.count_nonzero(inliers)
        if support > best_support:
            best_support = support
            best_inliers = inliers
            best_params  = (xc, yc, R)

    if best_params is None:
        raise RuntimeError("RANSAC (R tự do) không tìm được mô hình.")
    if best_support < max(int(min_inliers_frac * n), 6):
        raise RuntimeError(f"RANSAC tìm thấy {best_support} inliers (< ngưỡng).")

    return best_params, best_inliers

# ================== ĐỌC FILE (robust) ==================
df = pd.read_csv(
    file_path,
    header=None,
    names=['x','y','z'],
    sep=r'[,\s;]+',
    engine='python',
    usecols=[0,1,2],
    comment='#',
    skip_blank_lines=True,
    on_bad_lines='skip',
    dtype=str
)
for c in ['x','y','z']:
    df[c] = pd.to_numeric(df[c].str.strip(), errors='coerce')
df = df.dropna()

print(f"z range: [{df['z'].min():.6f}, {df['z'].max():.6f}] | total pts: {len(df)}")

# ================== LẬP DANH SÁCH CAO ĐỘ ==================
z_min, z_max = df['z'].min(), df['z'].max()
z_levels = np.arange(np.ceil((z_min)/z_step)*z_step, z_max + 1e-9, z_step)
# (nếu muốn neo vào giá trị cụ thể, có thể chỉnh điểm bắt đầu ở trên)

results = []   # mỗi phần tử: dict(z0, xc, yc, R, n_pts, n_inliers, cost, iters, success)

# ================== QUÉT TỪNG LÁT VÀ FIT ==================
for z0 in z_levels:
    mask = (df['z'] >= z0 - dz) & (df['z'] <= z0 + dz)
    pts  = df.loc[mask, ['x','y']].to_numpy(dtype=float)
    n_slice = len(pts)
    if n_slice < 30:
        results.append(dict(z0=z0, success=False, reason="slice too sparse", n_pts=n_slice))
        continue

    x, y = pts[:,0], pts[:,1]

    try:
        if R_known is not None:
            # R cố định
            (best_xc, best_yc), inliers_mask = ransac_circle_fixed_R(
                x, y, R_known,
                inlier_thresh=inlier_thresh,
                max_trials=ransac_max_trials,
                min_inliers_frac=min_inliers_frac,
                random_seed=random_seed
            )
            if do_lm_refine:
                xin, yin = x[inliers_mask], y[inliers_mask]
                xc, yc, R, cost, iters = lm_center_with_fixed_R(
                    xin, yin, best_xc, best_yc, R_known,
                    use_huber=use_huber, huber_c=huber_c,
                    max_iter=max_iter, lam0=lam0, tol_step=tol_step, tol_cost=tol_cost
                )
            else:
                xc, yc, R = best_xc, best_yc, R_known
                rho_tmp = np.sqrt((x[inliers_mask]-xc)**2 + (y[inliers_mask]-yc)**2)
                res_tmp = rho_tmp - R
                cost = np.sum(res_tmp**2); iters = 0
            n_inliers = int(np.count_nonzero(inliers_mask))
        else:
            # R tự do
            (best_xc, best_yc, best_R), inliers_mask = ransac_circle_variable_R(
                x, y,
                inlier_thresh=inlier_thresh,
                max_trials=ransac_max_trials,
                min_inliers_frac=min_inliers_frac,
                random_seed=random_seed
            )
            if do_lm_refine:
                xin, yin = x[inliers_mask], y[inliers_mask]
                # dùng R khởi tạo = best_R
                xc, yc, R, cost, iters = lm_center_radius(
                    xin, yin, best_xc, best_yc, best_R,
                    use_huber=use_huber, huber_c=huber_c,
                    max_iter=max_iter, lam0=lam0, tol_step=tol_step, tol_cost=tol_cost
                )
            else:
                xc, yc, R = best_xc, best_yc, best_R
                rho_tmp = np.sqrt((x[inliers_mask]-xc)**2 + (y[inliers_mask]-yc)**2)
                res_tmp = rho_tmp - R
                cost = np.sum(res_tmp**2); iters = 0
            n_inliers = int(np.count_nonzero(inliers_mask))

        results.append(dict(
            z0=z0, xc=xc, yc=yc, R=R,
            n_pts=n_slice, n_inliers=n_inliers,
            cost=float(cost), iters=int(iters),
            success=True
        ))

    except Exception as e:
        results.append(dict(z0=z0, success=False, reason=str(e), n_pts=n_slice))

# ================== IN BẢNG KẾT QUẢ ==================
print("\n=== KẾT QUẢ THEO CAO ĐỘ (mỗi 4 m) ===")
rows = []
for r in results:
    if r.get('success', False):
        rows.append([r['z0'], r['xc'], r['yc'], r['R'], r['n_pts'], r['n_inliers'], r['cost'], r['iters']])
    else:
        rows.append([r['z0'], np.nan, np.nan, np.nan, r['n_pts'], 0, np.nan, 0])
res_df = pd.DataFrame(rows, columns=['z0','xc','yc','R','n_pts','n_inliers','cost','iters'])
print(res_df.to_string(index=False, float_format=lambda v: f"{v:.6f}" if np.isfinite(v) else "nan"))

# ================== VẼ 2D TỪNG LÁT (chỉ lát thành công) ==================
for r in results:
    if not r.get('success', False):
        continue
    z0, xc, yc, R = r['z0'], r['xc'], r['yc'], r['R']
    mask = (df['z'] >= z0 - dz) & (df['z'] <= z0 + dz)
    sub = df.loc[mask, ['x','y']].to_numpy(dtype=float)
    x, y = sub[:,0], sub[:,1]
    rho = np.sqrt((x-xc)**2 + (y-yc)**2)
    res = np.abs(rho - R)
    inliers_mask = res <= inlier_thresh

    theta = np.linspace(0, 2*np.pi, 720)
    xcirc = xc + R*np.cos(theta)
    ycirc = yc + R*np.sin(theta)

    fig, ax = plt.subplots(1,1, figsize=(5.2,5.2))
    ax.scatter(x[~inliers_mask], y[~inliers_mask], s=1, alpha=0.25, label='outliers')
    ax.scatter(x[inliers_mask],  y[inliers_mask],  s=1, alpha=0.85, label='inliers')
    ax.plot(xcirc, ycirc, linewidth=2, label=f'circle (R={R:.4f} m)')
    ax.scatter([xc],[yc], s=25, label='center')
    ax.set_aspect('equal', 'box'); ax.grid(True); ax.legend(loc='best', fontsize=8)
    ax.set_title(f"Slice z={z0:.2f} ± {dz} m\ncenter=({xc:.3f},{yc:.3f}), R={R:.3f} m | inliers={int(inliers_mask.sum())}/{len(x)}")
    plt.show()

# ================== VẼ 3D: TẤT CẢ ĐƯỜNG TRÒN THEO CAO ĐỘ ==================
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401

fig = plt.figure(figsize=(7.5,6.5))
ax3d = fig.add_subplot(111, projection='3d')

# (tuỳ chọn) downsample đám mây để nền cho dễ nhìn
Nshow = 100000
if len(df) > Nshow:
    dff = df.sample(Nshow, random_state=0)
else:
    dff = df
ax3d.scatter(dff['x'], dff['y'], dff['z'], s=0.5, alpha=0.15, label='points')

# vẽ các đường tròn
for r in results:
    if not r.get('success', False):
        continue
    z0, xc, yc, R = r['z0'], r['xc'], r['yc'], r['R']
    theta = np.linspace(0, 2*np.pi, 360)
    xcirc = xc + R*np.cos(theta)
    ycirc = yc + R*np.sin(theta)
    zcirc = np.full_like(theta, z0)
    ax3d.plot(xcirc, ycirc, zcirc, linewidth=2)

ax3d.set_xlabel('X (m)'); ax3d.set_ylabel('Y (m)'); ax3d.set_zlabel('Z (m)')
ax3d.set_title('All fitted circles across elevations')
plt.tight_layout()
plt.show()
```

```python
# HDNC Outlier Filter (TEST MODE, no file write) for TXT input
# TXT format: whitespace-separated x y z per line (extra columns are ignored)

import open3d as o3d
import numpy as np
from pathlib import Path

# ===================== CONFIG =====================
INPUT_TXT   = "NP-05_SCN0001.txt"  # file .txt của bạn
VOXEL_SIZE  = 0.005         # 5 mm; 0 để bỏ qua
K_NEIGHBORS = 30            # 20–50
MAD_Z_TH    = 3.0           # >3 => mật độ thưa
SIGMA_TH    = 0.05          # surface variation
CONSIST_TH  = 0.75          # normal consistency
VOTE_REQ    = 2             # loại nếu >=2 tiêu chí xấu
SHOW_VIEWER = True         # True để bật viewer Open3D (mẫu nhỏ)
VIEW_SAMPLE = 100000        # số điểm tối đa khi xem (để nhẹ)
# ===================================================

def load_txt_xyz(path):
    pts = []
    with Path(path).open("r", encoding="utf-8", errors="ignore") as f:
        for line in f:
            s = line.strip()
            if not s: 
                continue
            # tách theo whitespace hoặc dấu phẩy
            parts = s.replace(",", " ").split()
            if len(parts) < 3:
                continue
            try:
                x, y, z = float(parts[0]), float(parts[1]), float(parts[2])
                pts.append((x, y, z))
            except:
                continue
    if not pts:
        raise ValueError("Không đọc được điểm nào từ TXT.")
    return np.asarray(pts, dtype=np.float64)

def voxel_downsample_np(points, voxel):
    if voxel is None or voxel <= 0:
        return points
    # lượng hoá theo kích thước voxel
    q = np.floor(points / voxel).astype(np.int64)
    # chọn 1 điểm đại diện mỗi voxel (hash bằng tuple)
    uniq, idx = np.unique(q, axis=0, return_index=True)
    return points[idx]

def build_kdtree_o3d(points):
    pcd = o3d.geometry.PointCloud(o3d.utility.Vector3dVector(points))
    tree = o3d.geometry.KDTreeFlann(pcd)
    return pcd, tree

def local_pca_features(points, idxs):
    P = points[idxs]                      # (k,3)
    ctr = P.mean(axis=0)
    Q = P - ctr
    C = (Q.T @ Q) / max(len(P), 1)        # 3x3 covariance
    w, v = np.linalg.eigh(C)              # trị riêng tăng dần
    tr = w.sum()
    lam1 = max(w[0], 0.0)
    sigma = lam1 / tr if tr > 0 else 1.0  # surface variation
    n = v[:, 0]
    n /= (np.linalg.norm(n) + 1e-12)
    mean_dist = float(np.linalg.norm(Q, axis=1).mean()) if len(Q) else np.inf
    return n, sigma, mean_dist

def mad_scores(x, eps=1e-9):
    med = np.median(x)
    mad = np.median(np.abs(x - med))
    z = np.abs(x - med) / (mad + eps)
    return z, med, mad

def hdncoutlier_filter(points, k=30, mad_z_th=3.0, sigma_th=0.05, consist_th=0.75, vote_req=2):
    N = len(points)
    pcd, tree = build_kdtree_o3d(points)

    normals = np.zeros((N,3), dtype=np.float64)
    sigma   = np.zeros(N, dtype=np.float64)
    mean_d  = np.zeros(N, dtype=np.float64)

    # Pass 1: PCA (normal, sigma, mean distance)
    k_eff = max(k, 3)
    for i in range(N):
        _, idxs, _ = tree.search_knn_vector_3d(points[i], k_eff)
        n_i, s_i, d_i = local_pca_features(points, idxs)
        normals[i] = n_i
        sigma[i]   = s_i
        mean_d[i]  = d_i

    # MAD density
    z, med_d, mad_d = mad_scores(mean_d)

    # Pass 2: normal consistency
    consist = np.zeros(N, dtype=np.float64)
    for i in range(N):
        _, idxs, _ = tree.search_knn_vector_3d(points[i], k_eff)
        # cosines với hàng xóm (dùng trị tuyệt đối để bỏ hướng)
        cosines = np.abs(normals[idxs] @ normals[i])
        consist[i] = float(cosines.mean()) if len(cosines) else 0.0

    # Quy tắc quyết định
    bad_density = z > mad_z_th
    bad_sigma   = sigma > sigma_th
    bad_normals = consist < consist_th

    votes = bad_density.astype(int) + bad_sigma.astype(int) + bad_normals.astype(int)
    is_out = votes >= vote_req
    keep_mask = ~is_out

    stats = {
        "N_in": int(N),
        "N_keep": int(np.count_nonzero(keep_mask)),
        "N_drop": int(np.count_nonzero(is_out)),
        "keep_ratio": float(np.count_nonzero(keep_mask) / max(N,1)),
        "median_mean_dist": float(med_d),
        "mad_mean_dist": float(mad_d),
        "sigma_mean": float(np.mean(sigma)),
        "consistency_mean": float(np.mean(consist)),
        "params": {
            "k": k, "mad_z_th": mad_z_th, "sigma_th": sigma_th,
            "consist_th": consist_th, "vote_req": vote_req,
            "voxel_size": VOXEL_SIZE
        }
    }
    return keep_mask, normals, sigma, z, consist, stats

def main():
    print("[1/4] Load TXT ...")
    pts = load_txt_xyz(INPUT_TXT)
    print(f"   - points (raw): {len(pts):,}")

    print("[2/4] Voxel downsample ...")
    pts_ds = voxel_downsample_np(pts, VOXEL_SIZE)
    print(f"   - points (voxel): {len(pts_ds):,}")

    print("[3/4] HDNC filtering (test mode) ...")
    keep_mask, normals, sigma, z, consist, stats = hdncoutlier_filter(
        pts_ds,
        k=K_NEIGHBORS,
        mad_z_th=MAD_Z_TH,
        sigma_th=SIGMA_TH,
        consist_th=CONSIST_TH,
        vote_req=VOTE_REQ
    )

    print("[4/4] Stats:")
    for k, v in stats.items():
        print(f"   {k}: {v}")

    # (Tuỳ chọn) hiển thị nhanh
    if SHOW_VIEWER:
        keep_pts = pts_ds[keep_mask]
        drop_pts = pts_ds[~keep_mask]
        # giảm để xem nhẹ
        if len(keep_pts) > VIEW_SAMPLE:
            keep_pts = keep_pts[np.random.choice(len(keep_pts), VIEW_SAMPLE, replace=False)]
        if len(drop_pts) > VIEW_SAMPLE:
            drop_pts = drop_pts[np.random.choice(len(drop_pts), VIEW_SAMPLE, replace=False)]
        p_keep = o3d.geometry.PointCloud(o3d.utility.Vector3dVector(keep_pts))
        p_drop = o3d.geometry.PointCloud(o3d.utility.Vector3dVector(drop_pts))
        p_keep.paint_uniform_color([0.1, 0.7, 0.1])  # xanh (giữ)
        p_drop.paint_uniform_color([0.9, 0.2, 0.2])  # đỏ (loại)
        o3d.visualization.draw_geometries([p_keep, p_drop])

if __name__ == "__main__":
    main()
```

```python
import open3d as o3d
import numpy as np
import os

# ==============================================================================
#                                  CẤU HÌNH
# ==============================================================================

# 1. SỬA ĐƯỜNG DẪN đến file .txt của bạn ở đây
FILE_PATH = r"NP-05_SCN0001.txt"

# 2. TINH CHỈNH CÁC THAM SỐ CỦA BỘ LỌC SOR
# ------------------------------------------------------------------------------
# nb_neighbors (k): Số lượng điểm lân cận để phân tích cho mỗi điểm.
# Giá trị lớn hơn sẽ làm cho thuật toán ít nhạy cảm hơn với nhiễu cục bộ.
# Giá trị phổ biến: 20-50
SOR_NEIGHBORS = 20

# std_ratio (alpha): Hệ số nhân độ lệch chuẩn.
# Quyết định mức độ "khắt khe" của bộ lọc. Các điểm có khoảng cách trung bình
# lớn hơn (trung bình chung + alpha * độ lệch chuẩn) sẽ bị loại bỏ.
# Giá trị nhỏ hơn sẽ lọc mạnh tay hơn.
# Giá trị phổ biến: 1.0 - 2.0
SOR_STD_RATIO = 2.0
# ------------------------------------------------------------------------------


def visualize_side_by_side(pcd_before, pcd_after):
    """
    Hàm helper để hiển thị hai đám mây điểm cạnh nhau để so sánh.
    """
    # Tô màu để dễ phân biệt
    pcd_before.paint_uniform_color([0.6, 0.6, 0.6])  # Màu xám cho dữ liệu gốc
    pcd_after.paint_uniform_color([0.0, 0.651, 0.929]) # Màu xanh cho dữ liệu đã lọc

    # Dịch chuyển đám mây điểm "sau" sang bên phải để hiển thị cạnh nhau
    # Lấy kích thước của đám mây gốc để biết cần dịch chuyển bao xa
    bbox = pcd_before.get_axis_aligned_bounding_box()
    translation_vector = np.array([bbox.get_max_bound()[0] - bbox.get_min_bound()[0] + 1.0, 0, 0])
    pcd_after.translate(translation_vector)

    # Hiển thị
    print("\nDisplaying visualization. Press 'q' in the window to close.")
    o3d.visualization.draw_geometries(
        [pcd_before, pcd_after],
        window_name="SOR Filter: Before (Left, Gray) vs After (Right, Blue)",
        width=1600,
        height=900
    )


def main():
    """
    Hàm chính để thực thi luồng xử lý.
    """
    print(f"Attempting to load point cloud from: {FILE_PATH}")
    if not os.path.exists(FILE_PATH):
        print(f"Error: File not found at '{FILE_PATH}'")
        return

    # === BƯỚC 1: ĐỌC DỮ LIỆU TỪ FILE TXT ===
    try:
        # Sử dụng NumPy để đọc file txt một cách hiệu quả
        points = np.loadtxt(FILE_PATH, delimiter=",") # Đổi delimiter=" " nếu các số cách nhau bởi dấu cách
        if points.shape[1] > 3:
             print(f"Warning: File has more than 3 columns. Using the first 3 columns as X, Y, Z.")
             points = points[:, :3] # Chỉ lấy 3 cột đầu tiên
        print(f"Successfully loaded {points.shape[0]} points.")
    except Exception as e:
        print(f"Error reading file: {e}")
        return

    if points.size == 0:
        print("Error: No points were loaded from the file.")
        return

    # Chuyển đổi mảng NumPy sang định dạng PointCloud của Open3D
    pcd_original = o3d.geometry.PointCloud()
    pcd_original.points = o3d.utility.Vector3dVector(points)


    # === BƯỚC 2: ÁP DỤNG BỘ LỌC STATISTICAL OUTLIER REMOVAL ===
    print("\nApplying Statistical Outlier Removal filter...")
    print(f"Parameters: nb_neighbors (k) = {SOR_NEIGHBORS}, std_ratio (alpha) = {SOR_STD_RATIO}")

    # remove_statistical_outlier trả về một tuple: (đám mây đã lọc, danh sách chỉ số của các điểm được giữ lại)
    pcd_filtered, ind = pcd_original.remove_statistical_outlier(
        nb_neighbors=SOR_NEIGHBORS,
        std_ratio=SOR_STD_RATIO
    )

    num_original = len(pcd_original.points)
    num_filtered = len(pcd_filtered.points)
    num_removed = num_original - num_filtered

    print("Filter applied successfully.")
    print(f" - Original number of points: {num_original}")
    print(f" - Points after filtering: {num_filtered}")
    print(f" - Outliers removed: {num_removed}")


    # === BƯỚC 3: TRỰC QUAN HÓA KẾT QUẢ ===
    # Tạo một bản sao của đám mây gốc để không làm thay đổi nó khi visualize
    pcd_before_vis = o3d.geometry.PointCloud(pcd_original)
    
    visualize_side_by_side(pcd_before_vis, pcd_filtered)


if __name__ == "__main__":
    main()
```

```python
import numpy as np

def circle_fit_by_taubin(points):
    """
    Khớp hình tròn bằng thuật toán Taubin, chuyển đổi từ code C++ của Nikolai Chernov.

    Args:
        points (np.array): Một mảng NumPy có hình dạng (N, 2), 
                           trong đó N là số điểm, các cột là X và Y.

    Returns:
        tuple: (a, b, r, iterations) 
               a, b: Tọa độ tâm hình tròn
               r: Bán kính hình tròn
               iterations: Số lần lặp của phương pháp Newton
    """
    n = points.shape[0]
    if n < 3:
        raise ValueError("Cần ít nhất 3 điểm để khớp hình tròn.")

    # --- 1. Tính toán giá trị trung bình và dịch chuyển dữ liệu ---
    mean_x = np.mean(points[:, 0])
    mean_y = np.mean(points[:, 1])
    
    # Dữ liệu đã được dịch chuyển (centered data)
    xi = points[:, 0] - mean_x
    yi = points[:, 1] - mean_y
    zi = xi**2 + yi**2

    # --- 2. Tính toán các Moment ---
    # np.mean() sẽ thực hiện cả việc cộng và chia cho n
    m_xx = np.mean(xi**2)
    m_yy = np.mean(yi**2)
    m_xy = np.mean(xi * yi)
    m_xz = np.mean(xi * zi)
    m_yz = np.mean(yi * zi)
    m_zz = np.mean(zi**2)

    # --- 3. Tính các hệ số của Phương Trình Đặc Trưng ---
    m_z = m_xx + m_yy
    cov_xy = m_xx * m_yy - m_xy**2
    var_z = m_zz - m_z**2
    
    a3 = 4 * m_z
    a2 = -3 * m_z**2 - m_zz
    a1 = var_z * m_z + 4 * cov_xy * m_z - m_xz**2 - m_yz**2
    a0 = m_xz * (m_xz * m_yy - m_yz * m_xy) + m_yz * (m_yz * m_xx - m_xz * m_xy) - var_z * cov_xy
    
    a22 = a2 + a2
    a33 = a3 + a3 + a3
    
    # --- 4. Tìm Nghiệm của Đa Thức (Newton's Method) ---
    ITER_MAX = 1000
    x = 0.0  # Bắt đầu từ x=0
    y = a0
    
    for i in range(ITER_MAX):
        # Tính đạo hàm P'(x)
        dy = a1 + x * (a22 + a33 * x)
        if dy == 0: # Tránh chia cho 0
            break

        x_new = x - y / dy
        
        # Kiểm tra điều kiện dừng
        if x_new == x or not np.isfinite(x_new):
            break
            
        # Tính giá trị mới của đa thức P(x_new)
        y_new = a0 + x_new * (a1 + x_new * (a2 + x_new * a3))
        
        if abs(y_new) >= abs(y):
            break
            
        x = x_new
        y = y_new
        
    iterations = i + 1

    # --- 5. Tính toán các tham số của hình tròn ---
    det = x**2 - x * m_z + cov_xy
    if det == 0:
        return None, None, None, iterations # Không thể giải
    
    # Tâm trong hệ tọa độ đã dịch chuyển
    x_center = (m_xz * (m_yy - x) - m_yz * m_xy) / (2 * det)
    y_center = (m_yz * (m_xx - x) - m_xz * m_xy) / (2 * det)

    # --- 6. Tổng hợp kết quả ---
    # Dịch chuyển tâm ngược lại về hệ tọa độ gốc
    a = x_center + mean_x
    b = y_center + mean_y
    r = np.sqrt(x_center**2 + y_center**2 + m_z)

    return a, b, r, iterations

# --- Ví dụ sử dụng ---
if __name__ == '__main__':

    x_points = slice_df['x']
    y_points = slice_df['y']
    
    data_points = np.c_[x_points, y_points]

    # Khớp hình tròn bằng thuật toán Taubin
    a_fit, b_fit, r_fit, iters = circle_fit_by_taubin(data_points)

    print("--- Kết quả khớp hình tròn bằng thuật toán Taubin ---")
    print(f"Giá trị khớp: a={a_fit:.4f}, b={b_fit:.4f}, r={r_fit:.4f}")
    print(f"Số lần lặp của Newton's method: {iters}")

```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import least_squares

# (Dán các hàm circle_fit_by_taubin và circle_residuals vào đây...)
# ...
def circle_fit_by_taubin(points):
    """Khớp hình tròn bằng thuật toán Taubin."""
    n = points.shape[0]
    if n < 3: return None, None, None, 0
    mean_x = np.mean(points[:, 0])
    mean_y = np.mean(points[:, 1])
    xi = points[:, 0] - mean_x
    yi = points[:, 1] - mean_y
    zi = xi**2 + yi**2
    m_xx = np.mean(xi**2); m_yy = np.mean(yi**2); m_xy = np.mean(xi * yi)
    m_xz = np.mean(xi * zi); m_yz = np.mean(yi * zi); m_zz = np.mean(zi**2)
    m_z = m_xx + m_yy; cov_xy = m_xx * m_yy - m_xy**2; var_z = m_zz - m_z**2
    a3 = 4 * m_z; a2 = -3 * m_z**2 - m_zz
    a1 = var_z * m_z + 4 * cov_xy * m_z - m_xz**2 - m_yz**2
    a0 = m_xz * (m_xz * m_yy - m_yz * m_xy) + m_yz * (m_yz * m_xx - m_xz * m_xy) - var_z * cov_xy
    a22 = a2 + a2; a33 = a3 + a3 + a3
    x = 0.0; y = a0; i = 0
    for i in range(99):
        dy = a1 + x * (a22 + a33 * x)
        if dy == 0: break
        x_new = x - y / dy
        if x_new == x or not np.isfinite(x_new): break
        y_new = a0 + x_new * (a1 + x_new * (a2 + x_new * a3))
        if abs(y_new) >= abs(y): break
        x = x_new; y = y_new
    iterations = i + 1
    det = x**2 - x * m_z + cov_xy
    if det == 0: return None, None, None, iterations
    x_center = (m_xz * (m_yy - x) - m_yz * m_xy) / (2 * det)
    y_center = (m_yz * (m_xx - x) - m_xz * m_xy) / (2 * det)
    a = x_center + mean_x; b = y_center + mean_y
    r = np.sqrt(x_center**2 + y_center**2 + m_z)
    return a, b, r, iterations

def circle_residuals(params, x_data, y_data):
    """Hàm tính vector phần dư hình học."""
    a, b, R = params
    return np.sqrt((x_data - a)**2 + (y_data - b)**2) - R

def circle_fit_by_lm(points, initial_guess):
    """Khớp hình tròn bằng Levenberg-Marquardt."""
    x_data = points[:, 0]
    y_data = points[:, 1]
    res = least_squares(circle_residuals, initial_guess, args=(x_data, y_data), method='lm')
    a_fit, b_fit, r_fit = res.x
    return a_fit, b_fit, r_fit

# ==============================================================================
# HÀM MỚI: TÍNH TOÁN CÁC CHỈ SỐ SAI SỐ
# ==============================================================================
def calculate_errors(points, circle_params):
    """
    Tính toán các chỉ số sai số cho kết quả khớp hình tròn.

    Args:
        points (np.array): Mảng điểm dữ liệu (N, 2).
        circle_params (tuple): Tham số của hình tròn đã khớp (a, b, R).

    Returns:
        dict: Một từ điển chứa các chỉ số sai số.
    """
    a, b, R = circle_params
    x_data = points[:, 0]
    y_data = points[:, 1]
    
    # 1. Tính toán vector phần dư (residuals)
    residuals = circle_residuals(circle_params, x_data, y_data)
    
    # 2. Tính MSE
    mse = np.mean(residuals**2)
    
    # 3. Tính RMSE (sigma)
    rmse = np.sqrt(mse)
    
    # 4. Tính MAE
    mae = np.mean(np.abs(residuals))
    
    # 5. (Tùy chọn) Tìm sai số lớn nhất để biết mức độ ảnh hưởng của outlier
    max_error = np.max(np.abs(residuals))
    
    error_metrics = {
        'residuals': residuals,
        'MSE': mse,
        'RMSE (sigma)': rmse,
        'MAE': mae,
        'Max Error': max_error,
        'Num Points': len(points)
    }
    
    return error_metrics


# ==============================================================================
# CẬP NHẬT QUY TRÌNH CHÍNH
# ==============================================================================
if __name__ == '__main__':
    # --- 1. Tạo dữ liệu mẫu ---

    data_points = np.c_[x_points, y_points]

    # --- 2. Chạy Taubin để có điểm khởi đầu ---
    print("--- Bước 1: Tìm điểm khởi đầu bằng Taubin Fit ---")
    a_taubin, b_taubin, r_taubin, _ = circle_fit_by_taubin(data_points)
    
    if a_taubin is not None:
        initial_guess = (a_taubin, b_taubin, r_taubin)
        print(f"Kết quả Taubin (Initial Guess): a={a_taubin:.4f}, b={b_taubin:.4f}, r={r_taubin:.4f}")

        # --- 3. Chạy Levenberg-Marquardt ---
        print("\n--- Bước 2: Tinh chỉnh bằng Levenberg-Marquardt Fit ---")
        a_lm, b_lm, r_lm = circle_fit_by_lm(data_points, initial_guess)
        final_params = (a_lm, b_lm, r_lm)
        print(f"Kết quả cuối cùng (LM): a={a_lm:.4f}, b={b_lm:.4f}, r={r_lm:.4f}")

        # --- 4. TÍNH TOÁN VÀ IN RA SAI SỐ ---
        print("\n--- Bước 3: Đánh giá sai số của kết quả cuối cùng ---")
        errors = calculate_errors(data_points, final_params)
        
        # In các chỉ số chính
        print(f"Số điểm dữ liệu: {errors['Num Points']}")
        print(f"Sai số bình phương trung bình gốc (RMSE): {errors['RMSE (sigma)']:.6f}")
        print(f"Sai số tuyệt đối trung bình (MAE): {errors['MAE']:.6f}")
        print(f"Sai số lớn nhất (gây ra bởi outlier): {errors['Max Error']:.4f}")
        
        # (Tùy chọn) In ra 5 phần dư đầu tiên
        # print("\n5 phần dư đầu tiên:")
        # print(errors['residuals'][:5])

        # --- 5. Vẽ đồ thị ---
        fig, ax = plt.subplots(1, figsize=(10, 10))
        ax.scatter(data_points[:, 0], data_points[:, 1], label='Data Points')
        circle_taubin = plt.Circle((a_taubin, b_taubin), r_taubin, color='orange', fill=False, linestyle='--', linewidth=2, label='Taubin Fit (Initial Guess)')
        ax.add_artist(circle_taubin)
        circle_lm = plt.Circle(final_params[:2], final_params[2], color='red', fill=False, linewidth=2, label='Levenberg-Marquardt Fit (Final)')
        ax.add_artist(circle_lm)
        ax.set_title('So sánh Taubin Fit và Levenberg-Marquardt Fit')
        ax.set_xlabel('Tọa độ X'); ax.set_ylabel('Tọa độ Y')
        ax.legend(); ax.grid(True); ax.set_aspect('equal', adjustable='box')
        plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import least_squares
import matplotlib.colors as mcolors

# ==============================================================================
# DÁN 5 HÀM ĐÃ VIẾT TRƯỚC ĐÓ VÀO ĐÂY
# 1. circle_fit_by_taubin(points)
# 2. circle_residuals(params, x_data, y_data)
# 3. circle_fit_by_lm(points, initial_guess)
# 4. calculate_errors(points, circle_params)
# 5. analyze_3d_tank(file_path, slice_thickness)
# ==============================================================================
# ... (dán code 5 hàm trên vào đây) ...
def circle_fit_by_taubin(points):
    """Khớp hình tròn bằng thuật toán Taubin."""
    n = points.shape[0]
    if n < 3: return None, None, None, 0
    mean_x = np.mean(points[:, 0])
    mean_y = np.mean(points[:, 1])
    xi = points[:, 0] - mean_x
    yi = points[:, 1] - mean_y
    zi = xi**2 + yi**2
    m_xx = np.mean(xi**2); m_yy = np.mean(yi**2); m_xy = np.mean(xi * yi)
    m_xz = np.mean(xi * zi); m_yz = np.mean(yi * zi); m_zz = np.mean(zi**2)
    m_z = m_xx + m_yy; cov_xy = m_xx * m_yy - m_xy**2; var_z = m_zz - m_z**2
    a3 = 4 * m_z; a2 = -3 * m_z**2 - m_zz
    a1 = var_z * m_z + 4 * cov_xy * m_z - m_xz**2 - m_yz**2
    a0 = m_xz * (m_xz * m_yy - m_yz * m_xy) + m_yz * (m_yz * m_xx - m_xz * m_xy) - var_z * cov_xy
    a22 = a2 + a2; a33 = a3 + a3 + a3
    x = 0.0; y = a0; i = 0
    for i in range(99):
        dy = a1 + x * (a22 + a33 * x)
        if dy == 0: break
        x_new = x - y / dy
        if x_new == x or not np.isfinite(x_new): break
        y_new = a0 + x_new * (a1 + x_new * (a2 + x_new * a3))
        if abs(y_new) >= abs(y): break
        x = x_new; y = y_new
    iterations = i + 1
    det = x**2 - x * m_z + cov_xy
    if det == 0: return None, None, None, iterations
    x_center = (m_xz * (m_yy - x) - m_yz * m_xy) / (2 * det)
    y_center = (m_yz * (m_xx - x) - m_xz * m_xy) / (2 * det)
    a = x_center + mean_x; b = y_center + mean_y
    r = np.sqrt(x_center**2 + y_center**2 + m_z)
    return a, b, r, iterations

def circle_residuals(params, x_data, y_data):
    """Hàm tính vector phần dư hình học."""
    a, b, R = params
    return np.sqrt((x_data - a)**2 + (y_data - b)**2) - R

def circle_fit_by_lm(points, initial_guess):
    """Khớp hình tròn bằng Levenberg-Marquardt."""
    x_data = points[:, 0]
    y_data = points[:, 1]
    res = least_squares(circle_residuals, initial_guess, args=(x_data, y_data), method='lm')
    a_fit, b_fit, r_fit = res.x
    return a_fit, b_fit, r_fit

def calculate_errors(points, circle_params):
    """Tính toán các chỉ số sai số cho kết quả khớp hình tròn."""
    a, b, R = circle_params
    residuals = circle_residuals(circle_params, points[:, 0], points[:, 1])
    mse = np.mean(residuals**2)
    rmse = np.sqrt(mse)
    mae = np.mean(np.abs(residuals))
    max_error = np.max(np.abs(residuals))
    error_metrics = {
        'residuals': residuals, 'MSE': mse, 'RMSE (sigma)': rmse,
        'MAE': mae, 'Max Error': max_error, 'Num Points': len(points)
    }
    return error_metrics

def analyze_3d_tank(file_path, slice_thickness=100):
    """Thực hiện quy trình cắt lớp và phân tích toàn bộ thùng phi 3D."""
    # ... (code hàm như trước) ...
    print("--- Bắt đầu quy trình phân tích thùng phi 3D ---")
    try:
        df = pd.read_csv(file_path, header=None, names=['x', 'y', 'z'], sep=',')
        print(f"Đã đọc thành công {len(df)} điểm từ file.")
    except FileNotFoundError:
        print(f"Lỗi: Không tìm thấy file tại '{file_path}'")
        return None, None
    z_min, z_max = df['z'].min(), df['z'].max()
    print(f"Phạm vi chiều cao của thùng phi: từ {z_min:.2f}m đến {z_max:.2f}m.")
    slice_edges = np.arange(z_min, z_max, slice_thickness)
    all_results = []
    print(f"\nBắt đầu xử lý {len(slice_edges)} lớp cắt với độ dày {slice_thickness}m...")
    for i, z_start in enumerate(slice_edges):
        z_end = z_start + slice_thickness
        slice_df = df[(df['z'] >= z_start) & (df['z'] < z_end)]
        if len(slice_df) < 10: continue
        z_mean = slice_df['z'].mean()
        data_points_2d = slice_df[['x', 'y']].values
        a_t, b_t, r_t, _ = circle_fit_by_taubin(data_points_2d)
        if a_t is None:
            print(f" - Lớp Z~{z_mean:.2f}m: Taubin fit thất bại, bỏ qua.")
            continue
        initial_guess = (a_t, b_t, r_t)
        a_lm, b_lm, r_lm = circle_fit_by_lm(data_points_2d, initial_guess)
        final_params = (a_lm, b_lm, r_lm)
        errors = calculate_errors(data_points_2d, final_params)
        all_results.append({
            'z_mean': z_mean, 'center_x': a_lm, 'center_y': b_lm,
            'radius': r_lm, 'rmse': errors['RMSE (sigma)'], 'mae': errors['MAE']
        })
        print(f" - Lớp Z~{z_mean:.2f}m: Đã xử lý. R = {r_lm:.4f}m, RMSE = {errors['RMSE (sigma)']:.4f}")
    if not all_results:
        print("Không có lớp nào được xử lý thành công.")
        return None, None
    results_df = pd.DataFrame(all_results)
    return df, results_df

# ==============================================================================
# LỚP MỚI: TẠO BÁO CÁO THEO MẪU
# ==============================================================================
class TankReportGenerator:
    def __init__(self, original_pointcloud_df, analysis_results_df):
        """
        Khởi tạo với dữ liệu gốc và kết quả phân tích.
        
        Args:
            original_pointcloud_df (pd.DataFrame): DataFrame chứa X, Y, Z của toàn bộ điểm.
            analysis_results_df (pd.DataFrame): DataFrame chứa kết quả phân tích từng lớp.
        """
        self.df_points = original_pointcloud_df
        self.df_results = analysis_results_df
        self.precompute_deviations()

    def precompute_deviations(self):
        """
        Tính toán trước sai số (deviation) cho mọi điểm trong point cloud.
        Đây là bước quan trọng để tạo các báo cáo sau này.
        """
        print("\nĐang tính toán sai số (deviation) cho toàn bộ đám mây điểm...")
        # Lấy tâm trung bình làm tham chiếu cho toàn bộ bồn
        mean_center_x = self.df_results['center_x'].mean()
        mean_center_y = self.df_results['center_y'].mean()
        
        # Tính góc và deviation cho mỗi điểm
        dx = self.df_points['x'] - mean_center_x
        dy = self.df_points['y'] - mean_center_y
        
        # atan2 cho kết quả từ -pi đến pi. Chuyển về 0 đến 2*pi và sang độ.
        self.df_points['angle'] = (np.arctan2(dy, dx) + 2 * np.pi) % (2 * np.pi) * 180 / np.pi
        
        # Nội suy bán kính lý tưởng tại cao độ Z của mỗi điểm
        # Dùng kết quả phân tích theo lớp để có bán kính chính xác tại từng độ cao
        ideal_radius_at_z = np.interp(self.df_points['z'], self.df_results['z_mean'], self.df_results['radius'])
        
        actual_distance = np.sqrt(dx**2 + dy**2)
        self.df_points['deviation'] = actual_distance - ideal_radius_at_z
        print(" -> Hoàn tất tính toán sai số.")

    def generate_buck_dents_report(self, z_bins=50, angle_bins=120, filename='report_buck_dents.png'):
        """Tạo báo cáo Buck-dents (Bản đồ nhiệt trải phẳng)."""
        print(f"\n1. Đang tạo báo cáo Buck-dents...")
        z_min, z_max = self.df_points['z'].min(), self.df_points['z'].max()
        
        # Tạo lưới 2D (chiều cao vs. góc) và tính deviation trung bình cho mỗi ô
        heatmap_data, z_edges, angle_edges = np.histogram2d(
            self.df_points['z'], self.df_points['angle'],
            bins=[z_bins, angle_bins],
            weights=self.df_points['deviation']
        )
        
        counts, _, _ = np.histogram2d(
            self.df_points['z'], self.df_points['angle'],
            bins=[z_bins, angle_bins]
        )
        
        # Tránh chia cho 0 ở những ô không có điểm
        heatmap_data = np.divide(heatmap_data, counts, out=np.zeros_like(heatmap_data), where=counts!=0)
        
        # Vẽ bản đồ nhiệt
        fig, ax = plt.subplots(figsize=(18, 6))
        # dùng T để ma trận khớp với trục
        im = ax.imshow(heatmap_data.T, origin='lower', aspect='auto', 
                         extent=[z_min, z_max, 0, 360], cmap='viridis')
        
        ax.set_title('Báo Cáo Biến Dạng Bề Mặt (Buck-dents Report)', fontsize=16)
        ax.set_xlabel('Cao trình Z (m)')
        ax.set_ylabel('Góc (độ)')
        fig.colorbar(im, ax=ax, label='Deviation (m)')
        plt.savefig(filename)
        plt.show()
        print(f" -> Đã lưu báo cáo tại '{filename}'")

    def generate_cross_section_report(self, target_z, slice_thickness=0.1, magnification=20, filename='report_cross_section.png'):
        """Tạo báo cáo mặt cắt ngang tại một cao độ Z cụ thể."""
        print(f"\n2. Đang tạo báo cáo mặt cắt ngang tại Z ~ {target_z}m...")
        lower = target_z - slice_thickness / 2
        upper = target_z + slice_thickness / 2
        slice_df = self.df_points[(self.df_points['z'] >= lower) & (self.df_points['z'] <= upper)]
        
        if len(slice_df) < 10:
            print(" -> Không đủ điểm tại cao độ này.")
            return

        # Tìm thông số vòng tròn cho lớp này từ kết quả đã có
        result_row = self.df_results.iloc[(self.df_results['z_mean'] - target_z).abs().argmin()]
        a, b, r = result_row['center_x'], result_row['center_y'], result_row['radius']
        
        fig, ax = plt.subplots(figsize=(10, 10))
        
        # Vẽ deviation plot được khuếch đại
        residuals = slice_df['deviation']
        plot_x = slice_df['x'] + residuals * np.cos(slice_df['angle'] * np.pi / 180) * magnification
        plot_y = slice_df['y'] + residuals * np.sin(slice_df['angle'] * np.pi / 180) * magnification
        ax.plot(plot_x, plot_y, color='black', linewidth=1, label=f'Deviation Plot (x{magnification})')
        
        # Vẽ hình tròn chuẩn
        circle = plt.Circle((a, b), r, color='red', fill=False, linewidth=2, label='Best-fit Radius')
        ax.add_artist(circle)
        
        # Tô màu cho deviation plot
        points = np.array([plot_x, plot_y]).T.reshape(-1, 1, 2)
        segments = np.concatenate([points[:-1], points[1:]], axis=1)
        norm = plt.Normalize(residuals.min(), residuals.max())
        lc = mcolors.LineCollection(segments, cmap='jet', norm=norm)
        lc.set_array(residuals)
        lc.set_linewidth(3)
        line = ax.add_collection(lc)
        fig.colorbar(line, ax=ax, label='Deviation (m)')

        ax.set_title(f'Báo Cáo Mặt Cắt Ngang tại Z ~ {target_z:.2f}m')
        ax.set_xlabel('Tọa độ X (m)')
        ax.set_ylabel('Tọa độ Y (m)')
        ax.set_aspect('equal', adjustable='box')
        ax.grid(True)
        plt.savefig(filename)
        plt.show()
        print(f" -> Đã lưu báo cáo tại '{filename}'")

    def generate_verticality_report(self, station_name="Station 5", filename='report_verticality.png'):
        """Tạo báo cáo mặt cắt dọc (độ thẳng đứng)."""
        print(f"\n3. Đang tạo báo cáo độ thẳng đứng...")
        fig, ax = plt.subplots(figsize=(8, 10))
        
        # Tính toán độ lệch của tâm so với trục trung bình
        mean_cx = self.df_results['center_x'].mean()
        center_deviation = self.df_results['center_x'] - mean_cx
        
        ax.plot(center_deviation, self.df_results['z_mean'])
        
        ax.set_title(f'Verticality Report - {station_name}')
        ax.set_xlabel('Độ lệch của tâm so với trục trung bình (m)')
        ax.set_ylabel('Cao trình Z (m)')
        ax.grid(True)
        ax.axvline(0, color='black', linestyle='--') # Vẽ trục trung tâm
        plt.savefig(filename)
        plt.show()
        print(f" -> Đã lưu báo cáo tại '{filename}'")

# ==============================================================================
# THỰC THI QUY TRÌNH HOÀN CHỈNH
# ==============================================================================
if __name__ == '__main__':
    # Đường dẫn file dữ liệu của bạn
    file_to_analyze = 'point_cloud_1.txt' # Thay bằng file thật của bạn
    
    # Bước 1: Phân tích dữ liệu 3D
    # Tạo dữ liệu mẫu nếu file không tồn tại
    try:
        pd.read_csv(file_to_analyze)
    except FileNotFoundError:
        print(f"Không tìm thấy file '{file_to_analyze}'. Đang tạo file dữ liệu mẫu...")
        R_ideal = 5.0; center_x, center_y = 10, 20
        z_coords = np.linspace(0, 10, 101)
        all_points_3d = []
        for z in z_coords:
            num_points_slice = 150
            angles = np.linspace(0, 2 * np.pi, num_points_slice)
            radius_at_z = R_ideal - 0.1 * np.cos(np.pi * z / 5)
            if 4.0 < z < 5.0: radius_at_z -= 0.2
            x = center_x + radius_at_z * np.cos(angles) + np.random.normal(0, 0.02, num_points_slice)
            y = center_y + radius_at_z * np.sin(angles) + np.random.normal(0, 0.02, num_points_slice)
            z_vals = np.full_like(x, z)
            for i in range(num_points_slice): all_points_3d.append([x[i], y[i], z_vals[i]])
        mock_data_3d = np.array(all_points_3d)
        np.savetxt(file_to_analyze, mock_data_3d, delimiter=",", fmt='%.4f')

    df_original, df_analysis = analyze_3d_tank(file_to_analyze, slice_thickness=1)

    # Bước 2: Tạo các báo cáo từ kết quả phân tích
    if df_original is not None and df_analysis is not None:
        report_generator = TankReportGenerator(df_original, df_analysis)
        
        # Tạo báo cáo 1: Buck-dents
        report_generator.generate_buck_dents_report()
        
        # Tạo báo cáo 2: Mặt cắt ngang (chọn một cao độ ở giữa để minh họa)
        middle_z = df_original['z'].median()
        report_generator.generate_cross_section_report(target_z=middle_z)
        
        # Tạo báo cáo 3: Độ thẳng đứng
        report_generator.generate_verticality_report(station_name="BUCKET E1")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import least_squares
import csv # Thư viện để làm việc với file CSV

# ==============================================================================
# DÁN CÁC HÀM VÀ LỚP ĐÃ VIẾT TRƯỚC ĐÓ VÀO ĐÂY
# ... (toàn bộ code từ các câu trả lời trước, bao gồm các hàm fit, analyze và lớp TankReportGenerator)
# ...
def circle_fit_by_taubin(points):
    """Khớp hình tròn bằng thuật toán Taubin."""
    n = points.shape[0]
    if n < 3: return None, None, None, 0
    mean_x = np.mean(points[:, 0])
    mean_y = np.mean(points[:, 1])
    xi = points[:, 0] - mean_x
    yi = points[:, 1] - mean_y
    zi = xi**2 + yi**2
    m_xx = np.mean(xi**2); m_yy = np.mean(yi**2); m_xy = np.mean(xi * yi)
    m_xz = np.mean(xi * zi); m_yz = np.mean(yi * zi); m_zz = np.mean(zi**2)
    m_z = m_xx + m_yy; cov_xy = m_xx * m_yy - m_xy**2; var_z = m_zz - m_z**2
    a3 = 4 * m_z; a2 = -3 * m_z**2 - m_zz
    a1 = var_z * m_z + 4 * cov_xy * m_z - m_xz**2 - m_yz**2
    a0 = m_xz * (m_xz * m_yy - m_yz * m_xy) + m_yz * (m_yz * m_xx - m_xz * m_xy) - var_z * cov_xy
    a22 = a2 + a2; a33 = a3 + a3 + a3
    x = 0.0; y = a0; i = 0
    for i in range(99):
        dy = a1 + x * (a22 + a33 * x)
        if dy == 0: break
        x_new = x - y / dy
        if x_new == x or not np.isfinite(x_new): break
        y_new = a0 + x_new * (a1 + x_new * (a2 + x_new * a3))
        if abs(y_new) >= abs(y): break
        x = x_new; y = y_new
    iterations = i + 1
    det = x**2 - x * m_z + cov_xy
    if det == 0: return None, None, None, iterations
    x_center = (m_xz * (m_yy - x) - m_yz * m_xy) / (2 * det)
    y_center = (m_yz * (m_xx - x) - m_xz * m_xy) / (2 * det)
    a = x_center + mean_x; b = y_center + mean_y
    r = np.sqrt(x_center**2 + y_center**2 + m_z)
    return a, b, r, iterations

def circle_residuals(params, x_data, y_data):
    """Hàm tính vector phần dư hình học."""
    a, b, R = params
    return np.sqrt((x_data - a)**2 + (y_data - b)**2) - R

def circle_fit_by_lm(points, initial_guess):
    """Khớp hình tròn bằng Levenberg-Marquardt."""
    x_data = points[:, 0]
    y_data = points[:, 1]
    res = least_squares(circle_residuals, initial_guess, args=(x_data, y_data), method='lm')
    a_fit, b_fit, r_fit = res.x
    return a_fit, b_fit, r_fit

def calculate_errors(points, circle_params):
    """Tính toán các chỉ số sai số cho kết quả khớp hình tròn."""
    a, b, R = circle_params
    residuals = circle_residuals(circle_params, points[:, 0], points[:, 1])
    mse = np.mean(residuals**2)
    rmse = np.sqrt(mse)
    mae = np.mean(np.abs(residuals))
    max_error = np.max(np.abs(residuals))
    error_metrics = {
        'residuals': residuals, 'MSE': mse, 'RMSE (sigma)': rmse,
        'MAE': mae, 'Max Error': max_error, 'Num Points': len(points)
    }
    return error_metrics

def analyze_3d_tank(file_path, slice_thickness=0.1):
    """Thực hiện quy trình cắt lớp và phân tích toàn bộ thùng phi 3D."""
    # ... (code hàm như trước) ...
    print("--- Bắt đầu quy trình phân tích thùng phi 3D ---")
    try:
        df = pd.read_csv(file_path, header=None, names=['x', 'y', 'z'], sep=',')
        print(f"Đã đọc thành công {len(df)} điểm từ file.")
    except FileNotFoundError:
        print(f"Lỗi: Không tìm thấy file tại '{file_path}'")
        return None, None
    z_min, z_max = df['z'].min(), df['z'].max()
    print(f"Phạm vi chiều cao của thùng phi: từ {z_min:.2f}m đến {z_max:.2f}m.")
    slice_edges = np.arange(z_min, z_max, slice_thickness)
    all_results = []
    print(f"\nBắt đầu xử lý {len(slice_edges)} lớp cắt với độ dày {slice_thickness}m...")
    for i, z_start in enumerate(slice_edges):
        z_end = z_start + slice_thickness
        slice_df = df[(df['z'] >= z_start) & (df['z'] < z_end)]
        if len(slice_df) < 10: continue
        z_mean = slice_df['z'].mean()
        data_points_2d = slice_df[['x', 'y']].values
        a_t, b_t, r_t, _ = circle_fit_by_taubin(data_points_2d)
        if a_t is None:
            print(f" - Lớp Z~{z_mean:.2f}m: Taubin fit thất bại, bỏ qua.")
            continue
        initial_guess = (a_t, b_t, r_t)
        a_lm, b_lm, r_lm = circle_fit_by_lm(data_points_2d, initial_guess)
        final_params = (a_lm, b_lm, r_lm)
        errors = calculate_errors(data_points_2d, final_params)
        all_results.append({
            'z_mean': z_mean, 'center_x': a_lm, 'center_y': b_lm,
            'radius': r_lm, 'rmse': errors['RMSE (sigma)'], 'mae': errors['MAE'],
            'slice_points': data_points_2d
        })
        print(f" - Lớp Z~{z_mean:.2f}m: Đã xử lý. R = {r_lm:.4f}m, RMSE = {errors['RMSE (sigma)']:.4f}")
    if not all_results:
        print("Không có lớp nào được xử lý thành công.")
        return None, None
    results_df = pd.DataFrame(all_results)
    return df, results_df

class TankReportGenerator:
    # ... (Dán các hàm __init__, precompute_deviations, và các hàm generate_* khác vào đây)
    def __init__(self, original_pointcloud_df, analysis_results_df):
        self.df_points = original_pointcloud_df
        self.df_results = analysis_results_df
        self.precompute_deviations()
        
    def precompute_deviations(self):
      """Tính toán sai số cho toàn bộ point cloud."""
      # ... (code như cũ)
      pass
      
    def generate_buck_dents_report(self, z_bins=50, angle_bins=120, filename='report_buck_dents.png'):
      """Tạo báo cáo Buck-dents."""
      # ... (code như cũ)
      pass
      
    # ... (các hàm generate khác)
    #==========================================================================
    # HÀM MỚI ĐỂ XUẤT FILE CSV
    #==========================================================================
    def generate_csv_report(self, 
                              filename='report_ovality_circumference.csv',
                              jacket_name='Jacket-01',
                              nominal_radius=None,
                              oor_tolerance=0.007,
                              circum_tolerance=0.020):
        """
        Tạo báo cáo dạng CSV theo mẫu của Ørsted.

        Args:
            filename (str): Tên file CSV đầu ra.
            jacket_name (str): Tên của cấu kiện.
            nominal_radius (float): Bán kính danh nghĩa (từ bản vẽ) để tính độ lệch chu vi.
            oor_tolerance (float): Ngưỡng dung sai cho phép đối với độ vênh (OOR).
            circum_tolerance (float): Ngưỡng dung sai cho phép đối với độ lệch chu vi.
        """
        print(f"\n4. Đang tạo báo cáo CSV '{filename}'...")

        # --- Bước 1: Chuẩn bị các giá trị tính toán ---
        
        # Nếu không có bán kính danh nghĩa, dùng bán kính trung bình
        if nominal_radius is None:
            print("Cảnh báo: Không có bán kính danh nghĩa. Sử dụng bán kính trung bình làm tham chiếu.")
            nominal_radius = self.df_results['radius'].mean()
        
        nominal_circumference = 2 * np.pi * nominal_radius
        
        # Tạo một DataFrame mới để chứa kết quả cho báo cáo
        report_data = []
        for index, row in self.df_results.iterrows():
            z = row['z_mean']
            a, b, r = row['center_x'], row['center_y'], row['radius']
            points = row['slice_points']
            
            # Tính Dmax, Dmin một cách bền vững
            residuals = circle_residuals((a, b, r), points[:, 0], points[:, 1])
            d_min = 2 * (r + np.percentile(residuals, 1)) # Bỏ 1% giá trị nhỏ nhất
            d_max = 2 * (r + np.percentile(residuals, 99)) # Bỏ 1% giá trị lớn nhất
            
            # Tính toán các giá trị cho báo cáo
            oor = d_max - d_min
            circumference = 2 * np.pi * r
            circum_deviation = circumference - nominal_circumference
            
            # Kiểm tra Accepted/Rejected
            is_oor_ok = "Accepted" if oor <= oor_tolerance else "Rejected"
            is_circum_ok = "Accepted" if abs(circum_deviation) <= circum_tolerance else "Rejected"
            final_status = "Accepted" if (is_oor_ok == "Accepted" and is_circum_ok == "Accepted") else "Rejected"
            
            report_data.append({
                'Survey pos.': f"+{z*1000:.0f} mm", # Chuyển m sang mm
                'Dmax': f"{d_max:.5f}",
                'Dmin': f"{d_min:.5f}",
                'Deviation (OOR)': f"{oor:.5f}",
                'Circumference': f"{circumference:.5f}",
                'Deviation (Circumference)': f"{circum_deviation:.5f}",
                'Accepted / Rejected': final_status
            })
            
        df_report = pd.DataFrame(report_data)

        # --- Bước 2: Ghi file CSV theo đúng định dạng ---
        with open(filename, 'w', newline='', encoding='utf-8') as f:
            writer = csv.writer(f)
            
            # Ghi các dòng header
            writer.writerow(['Ørsted Wind Power A/S'])
            writer.writerow(['Contractor', '', 'Project'])
            writer.writerow(['', '', '', 'Revision:'])
            writer.writerow(['', '', '', 'Date:'])
            writer.writerow(['', '', '', 'Page 1 of 1'])
            writer.writerow(['Bucket Ovality & Circumference'])
            writer.writerow(['Jacket Name', jacket_name, 'Temperature', '', 'Calibration date'])
            writer.writerow(['Date', '', 'Measurement equipment', '', 'Inspected by'])
            writer.writerow(['Time', '', 'Serial no.', '', 'Measurement procedure rev no.'])
            writer.writerow(['Remarks'])
            writer.writerow(['"Jacket to be surveyed according to requirements in OTS-F1150.'])
            writer.writerow(['Survey must document the ovality and circumfernce of the buckets"'])
            # Ghi các dòng trống
            for _ in range(15):
                writer.writerow([])
            
            # Ghi phần kết quả
            writer.writerow(['Survey results'])
            writer.writerow(df_report.columns.tolist()) # Ghi header của bảng
            writer.writerows(df_report.values.tolist()) # Ghi dữ liệu

        print(f" -> Đã lưu báo cáo thành công tại '{filename}'")
        return df_report

# ==============================================================================
# THỰC THI QUY TRÌNH HOÀN CHỈNH
# ==============================================================================
if __name__ == '__main__':
    # Đường dẫn file dữ liệu của bạn
    file_to_analyze = 'tank_3d_data.csv' 
    
    # Tạo dữ liệu mẫu nếu cần
    try: pd.read_csv(file_to_analyze)
    except FileNotFoundError:
        # ... (code tạo dữ liệu mẫu như cũ)
        pass

    # Bước 1: Phân tích dữ liệu 3D
    df_original, df_analysis = analyze_3d_tank(file_to_analyze, slice_thickness=0.1)

    # Bước 2: Tạo các báo cáo
    if df_original is not None and df_analysis is not None:
        report_generator = TankReportGenerator(df_original, df_analysis)
        
        # ... (gọi các hàm generate báo cáo hình ảnh nếu muốn)

        # Tạo báo cáo CSV
        # CUNG CẤP BÁN KÍNH DANH NGHĨA (THEO THIẾT KẾ) Ở ĐÂY (tính bằng mét)
        NOMINAL_RADIUS_FROM_DESIGN = 5.0 
        
        csv_report_df = report_generator.generate_csv_report(
            filename='Bucket_E1_Ovality_Report.csv',
            jacket_name='BUCKET E1',
            nominal_radius=NOMINAL_RADIUS_FROM_DESIGN,
            oor_tolerance=0.015,  # Dung sai OOR là 15mm
            circum_tolerance=0.025 # Dung sai chu vi là 25mm
        )
        
        print("\n--- Xem trước báo cáo CSV ---")
        print(csv_report_df.to_string())
```
